```python
# Verificacion de nulos
df.isnull().sum()
```

```python
# Calcular porcentaje de valores nulos por columna
null_percentages = pd.DataFrame({
    'Columna': df.columns,
    'Valores Nulos': df.isnull().sum().values,
    'Porcentaje Nulos (%)': (df.isnull().sum().values / len(df) * 100).round(2)
})

# Ordenar de mayor a menor porcentaje
null_percentages = null_percentages.sort_values('Porcentaje Nulos (%)', ascending=False)

# Mostrar resultado
print("PORCENTAJE DE VALORES NULOS POR COLUMNA \n")
print(null_percentages.to_string(index=False))
```

```python
# verificacion de duplicados
duplicados_totales = df.duplicated().sum()
print(f"Filas duplicadas\n")
print(f"Numero total de filas duplicadas: {duplicados_totales}")
print(f"Porcentaje de duplicados: {(duplicados_totales/len(df)*100):.2f}%")
```

```python
# Busca anomalias
df.describe()
```

```python
# Convertir la columna Date a formato datetime
fechas_convertidas = pd.to_datetime(df['Date'], errors='coerce')

# Contar fechas invalidas
fechas_invalidas = fechas_convertidas.isna().sum()

print(f"\nFechas con formato invalido: {fechas_invalidas}")

# Porcentaje de fechas invalidas
print(f"Porcentaje de fechas invalidas: {(fechas_invalidas/len(df)*100):.2f}%")

# Rango de fechas validas
print(f"\nRango de fechas validas:")
print(f"Fecha mas antigua: {fechas_convertidas.min()}")
print(f"Fecha mas reciente: {fechas_convertidas.max()}")
```
